```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2hsv

rgb_img = io.imread("https://siapdok.id/wp-content/uploads/2022/06/katarak-Mata.png")

hsv_hsv = rgb2hsv(rgb_img)
hue_img = hsv_hsv[:, :, 2]

fig, ([ax0, ax1]) = plt.subplots(1, 2, figsize=(20, 20))

ax0.imshow(hue_img)
ax0.set_title("Mata Katarak (Original)")
ax1.imshow(hsv_hsv)
ax1.set_title("Mata Katarak (Grayscale)")

plt.show()
```

```python
from skimage import data, filters, io
import matplotlib.pyplot as plt
import cv2

img = io.imread("https://mhomecare.co.id/blog/wp-content/uploads/2020/07/Katarak.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img1= cv2.Canny(img, 20, 40)

imgR = filters.roberts(img)
imgP = filters.prewitt(img)
fig, ax = plt.subplots(1, 3, figsize=(10, 8))
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Image Asli")

ax[1].imshow(img1, cmap='gray')
ax[1].set_title("Image Menggunakan Canny")

ax[2].imshow(imgP, cmap='gray')
ax[2].set_title("Image Menggunakan Prewitt")

plt.show()
```

```python
import cv2
import numpy as np
from skimage import io

image=io.imread("https://mhomecare.co.id/blog/wp-content/uploads/2020/07/Katarak.jpg")
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
io.imshow(image)

ret, thresh = cv2.threshold(gray, 45, 255, 0)
contours, hierarchy=cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
n = len(contours)-1
contours=sorted(contours, key=cv2.contourArea, reverse=False)[:n]
for c in contours:
  hull=cv2.convexHull(c)
cv2.drawContours(image,[hull],0,(0,255,0),2)
io.imshow(image)
io.show()
```

```python
import cv2
import numpy as np
from skimage import io

image=io.imread("https://mhomecare.co.id/blog/wp-content/uploads/2020/07/Katarak.jpg")
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
io.imshow(image)

ret, thresh=cv2.threshold(gray, 55, 255, 60)
contours, hierarchy=cv2.findContours(thresh.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
n = len(contours)-1
contours=sorted(contours, key=cv2.contourArea, reverse=False)[:n]
for c in contours:
  hull=cv2.convexHull(c)
cv2.drawContours(image,[hull],0,(0,255,0),2)
io.imshow(image)
io.show()
```

```python
import numpy as np
import cv2

#memanggil Library cv2 yaitu OpenCv
import matplotlib.pyplot as plt
from skimage import io

#Membaca Link Gambar
img = io.imread("https://mhomecare.co.id/blog/wp-content/uploads/2020/07/Katarak.jpg")

#Mengubah Gambar BGR ke HSV
grid_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

#Menentukan Nilai Batas
lower = np.array([0, 100, 100])
upper = np.array([4000, 255, 255])

#Hasil Segmentasi
mask = cv2.inRange(grid_HSV, lower, upper)
res = cv2.bitwise_and(img, grid_HSV, mask=mask)

fig, ax = plt.subplots(1, 4, figsize=(20, 8))
ax[0].imshow(img)
ax[0].set_title("Image Asli")
ax[1].imshow(grid_HSV)
ax[1].set_title("Image Grid_HSV")
ax[2].imshow(mask)
ax[2].set_title("Image Range")
ax[3].imshow(res)
ax[3].set_title("Image Bitwise")
plt.show()
```
